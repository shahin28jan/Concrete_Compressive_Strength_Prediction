```python
# plot for categorical columns
plt.figure(figsize=(15, 15))
#plt.suptitle('First Chart of bivariate Analysis of data', fontsize=20, fontweight='bold', alpha=0.8,y=1)
cols = ['Cement','Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age', 'Concrete compressive strength']
for i in range(0, len(cols)):
    plt.subplot(3, 3, i+1)
    sns.distplot(df[cols[i]])
    #sns.barplot(data = df, x = df[cols[i]], y = 'Concrete compressive strength')
    #plt.xlabel(df[i])
    plt.tight_layout()
```

```python
# plot for categorical columns
plt.figure(figsize=(15, 15))
#plt.suptitle('First Chart of bivariate Analysis of data', fontsize=20, fontweight='bold', alpha=0.8,y=1)
cols = ['Cement','Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age', 'Concrete compressive strength']
for i in range(0, len(cols)):
    plt.subplot(5, 2, i+1)
    #sns.distplot(df[cols[i]])
    sns.barplot(data = df, x = df[cols[i]], y = 'Concrete compressive strength')
    #plt.xlabel(df[i])
    plt.tight_layout()
```
